```python
import pandas as pd
import numpy as np
```

```python
#Arithmatic & Data Allignment
```

```python
df1 = pd.DataFrame(np.arange(9).reshape((3,3)),
                  index=['A','B','C'],
                  columns=list('bcd'))
df2 = pd.DataFrame(np.arange(16).reshape((4,4)),
                  index=['A','B','C','D'],
                  columns=list('bcyz'))

display(df1,df2)
df3 = df1 + df2
display(df3)
```

```python
df3.loc['A','b'] = np.nan
display(df3)
```

```python
display(df3.loc[['B','C'],['b','c']])
```

```python
display(df3.iloc[:,1:])
```

```python
display(df3.iloc[[1,2],[0,1]])
```
